```python
from bs4 import BeautifulSoup
import requests
import numpy as np
```

```python
baseurl="https://fruitboxco.com/"
```

```python
headers= {
    'User-Agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.157 Safari/537.36"
}
```

```python
productlinks=[]
```

```python
for x in range (1,3):
        r= requests.get(f"https://fruitboxco.com/collections/daily-fruits?page={x}" )
        soup=BeautifulSoup(r.content, 'lxml')
        productlist= soup.find_all('h2',class_="ProductItem__Title Heading")
        productlist
        for i in productlist:
            for link in i.find_all('a', href=True):
                productlinks.append(baseurl+link['href'])
len(productlinks)
```

```python
fruits=[]
for links in productlinks:
    r = requests.get(links, headers=headers)
    soup=BeautifulSoup(r.content, 'lxml')
    price=soup.find('span',class_="ProductMeta__Price Price Price--highlight Text--subdued u-h4").text.strip()
    name=soup.find('h1', class_="ProductMeta__Title Heading u-h2").text.strip()
    freshfruits={'Name':name, 'Price': price}
    fruits.append(freshfruits)
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(fruits)
df
```

```python
df.to_excel('fruitbox.xlsx')
```
